```python
import nltk;
from nltk.tokenize import RegexpTokenizer, word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
nltk.download('stopwords')
```

```python
dataframe=pd.read_csv('/content/drive/MyDrive/spam.csv',encoding='ISO-8859-1')
le=LabelEncoder();
```

```python
data=dataframe.to_numpy()
print(data[0])
print(data[2])
```

```python
X=data[:,1]#data in the 1st col ie the message
y=data[:,0]#data in the second column ie the spam/non-spam
```

```python
tokenizer=RegexpTokenizer('\w+')
stop_words=set(stopwords.words('english'))
ps=PorterStemmer()
```

```python
def rem_words_and_stem(data):
  data=data.lower()
  words=tokenizer.tokenize(data)
  removed_stopwords=[w for w in words if w not in stop_words]
  stemmed_words=[ps.stem(w) for w in removed_stopwords]
  clean_text=' '.join(stemmed_words)
  return clean_text
```

```python
def cleandata(document):
  d=[]
  for i in document:
    d.append(rem_words_and_stem(i))
  return d
```

```python
cleandocument=cleandata(X)
print(cleandocument[:10])
```

```python
cv=CountVectorizer()
```

```python
#create the library of our words from the training set
vc=cv.fit_transform(cleandocument)
print(vc)
print(cv.vocabulary_)#set of our vocabulary that we have created from our data set
```

```python
X=vc.todense()
print(X)
```
